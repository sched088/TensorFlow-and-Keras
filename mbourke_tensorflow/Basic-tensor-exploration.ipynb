### Getting information from tensors

* Shape
* Rank
* Axis / dimension
* Size

```python
# Create a rank 4 tensor (4 dimensions)
rank_4_tensor = tf.zeros(shape=(2,3,4,5))
rank_4_tensor
```

```python
from tensorflow.python.ops.gen_array_ops import rank
# Get various attributes of our tensor:

print("Datatype of every element:", rank_4_tensor.dtype)
print("Number of dimensions (rank):", rank_4_tensor.ndim)
print("Shape of tensor:", rank_4_tensor.shape)
print("Elements along the 0 axis:", rank_4_tensor.shape[0])
print("Total number of elements in our tensor:", tf.size(rank_4_tensor))
print("Total number of elements in our tensor:", tf.size(rank_4_tensor).numpy())
                    
```

### Manipulating Tensors (tensor operations)

**basic operations**

`+`,`-`,`*`,`/`

```python
# You can add values to a tensor using the addition operator
tensor = tf.constant([[10,7],[3,4]])
tensor + 10
```

```python
# Note: Original tensor is unchanged
tensor
```

```python
# multiplication
tensor * 10
```

```python
# subtraction
tensor - 10
```

```python
# We can use the tensorflow built-in function too
# for high speed on gpu use tf.math built-in functions
tf.multiply(tensor, 10)
```

**Matrix Multiplication**

One of the most common operations in machine learning/tensor operations. 

There are two rules that our tensors/matrices need to fulfil to be multiplied

1. The inner dimensions must match
2. The reuslting matrix has the shape of the inner dimensions

```python
# matrix multiplication in tensorflow
tf.matmul(tensor, tensor)
```

```python
tensor * tensor
```

```python
# Matrix multiplication with Python operator '@'

tensor @ tensor
```

```python
#  Create (3,2) tensor
X = tf.constant([[1,2],
                 [3,4],
                 [5,6]])
Y = tf.constant(np.arange(7,13).reshape(3,2), dtype=tf.int32)

X, Y
```

```python
# Matrix multiply tensors of same shape
X @ Y
```

```python
# Need them to be inverse(?) shapes
X @ tf.transpose(Y)
```

```python
# see transpose of Y
tf.transpose(Y)
```

```python
# change the shape of Y
tf.reshape(Y, shape=(2,3))
```

```python
# Try to multiply X by reshaped Y

X @ tf.reshape(Y, shape=(2,3))
```

## Dot Product

Two ways of performing matrix multiplication

* `tf.matmul()`
* `tf.tensordot()`

```python
# perform the dot produict on X and Y (requires a transpose)
tf.tensordot(X, tf.transpose(Y), axes=1)
```

```python
# perform matrix mult on X and Y (transposed)
tf.matmul(X, tf.transpose(Y))
```

```python
# perform matrix mult between X and Y (reshaped)
tf.matmul(X, tf.reshape(Y, shape=(2,3)))
```

```python
#  check the values of Y, reshape Y and transpose Y
print("normal Y:")
print(Y,"\n")

print("Y reshaped to (2,3):")
print(tf.reshape(Y, (2,3), "\n"))

print("Y transposed:")
print(tf.transpose(Y))
```

### Aggregating Tensors

Condensing tensors from multiple values down to a smaller amount of values

```python
# Get the absolute values
D = tf.constant([-7, -10])
D 
```

```python
tf.abs(D)
```

Different forms of aggregation
* min
* max
* mean
* sum

```python
tf.reduce_min(D) # minimum
```

```python
tf.reduce_max(D) # maximum
```

```python
tf.reduce_mean(D) # mean
```

```python
tf.reduce_sum(D) # sum
```

```python
tf.reduce_prod(D) #product of scalars
```

```python
import tensorflow_probability as tfp
```

```python
tfp.stats.variance(D) # variance
```

```python
tf.math.reduce_variance(tf.cast(D, dtype=tf.float32))
```

```python
tf.math.reduce_std(tf.cast(D, dtype=tf.float32))
```

### Finding positional maximum and minimum

```python
# create new tensor to find positional max, min
tf.random.set_seed(42)
F = tf.random.uniform(shape=[50])
F
```

```python
# find the positional maximum -- argmax
tf.argmax(F)
```

```python
F[tf.argmax(F)] # returns the max value of the tensor
```

```python
tf.reduce_max(F) # also returns the max value of the tensor
```

### TensorFlow and NumPy
TensorFlow interacts easily with NumPy arrays
* allows easy access to numpy functionality

```python
# create tensor directly from numpy array
J = tf.constant(np.array([3.,7.,10.]))
J
```

```python
# convert tensor back to NumPy array
np.array(J), type(np.array(J))
```

```python
J.numpy(), type(J.numpy())
```

```python
# watch for data types when creating tensors from arrays!
numpy_J = tf.constant(np.array([3.,7.,10.]))
tensor_J = tf.constant([3.,7.,10.])

# Notice the data types are different!
numpy_J.dtype, tensor_J.dtype
```
